I approached this assignment by first completing several Suduko puzzles and trying to figure out a process. Then I researched how Suduko puzzles are typically solved and found a name for the process called "cross-hatching." Cross-hatching starts by identifying the numbers missing from the first box in the puzzle. Then one by one, you would check to see if you could place the numbers from the missing numbers list in any spot. 

image.png

So in the case for the first box, the missing numbers are [2, 3, 5, 7, 9]. You would first look to see if you could place the 2 without violating any rules and without guessing. You couldn't. You would move on to the 3. Then the 5 and so on. For this first box, you can place the 9 in the spot above the 8.
